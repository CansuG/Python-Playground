```python
import cv2
```

```python
### load image
img = cv2.imread('/content/cat-image.jpeg')
# Convert the image to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
### Display image on google colab with OpenCV
from google.colab.patches import cv2_imshow
cv2_imshow(gray_image)
```

```python
### 1.Flip the image vertically

# Get the image dimensions
height, width = gray_image.shape

# Create a new image of the same size with blank pixels
flipped = [[0 for i in range(width)] for j in range(height)]

# Flip the image vertically
for i in range(height):
    for j in range(width):
        flipped[height-i-1][j] = gray_image[i][j]

# Display the flipped image
import numpy as np
cv2_imshow(np.array(flipped, dtype=np.uint8))
```

```python
### 2.Flip the image horizontally

# Create a new image of the same size with blank pixels
flipped = [[0 for i in range(width)] for j in range(height)]

# Flip the image horizontally
for i in range(height):
    for j in range(width):
        flipped[i][width-1-j] = gray_image[i][j]

# Display the flipped image
cv2_imshow(np.array(flipped, dtype=np.uint8))
```

```python
### 3.Rotate the image 90 degrees clockwise

# Create a new image of the same size with blank pixels
rotated = [[0 for i in range(height)] for j in range(width)]

# Rotate the image by 90 degrees clockwise
for i in range(height):
    for j in range(width):
        rotated[j][height-1-i] = gray_image[i][j]

# Display the rotated image
cv2_imshow(np.array(rotated, dtype=np.uint8))
```

```python
### 4.Rotate the image 90 degrees counterclockwise

# Create a new image of the same size with blank pixels
rotated = [[0 for i in range(height)] for j in range(width)]

# Rotate the image by 90 degrees clockwise
for i in range(height):
    for j in range(width):
        rotated[width-1-j][i] = gray_image[i][j]

# Display the rotated image
cv2_imshow(np.array(rotated, dtype=np.uint8))
```

```python
### 5.Resize the image to half by keeping the aspect ratio

# Define the new dimensions of the resized image
new_height = height // 2
new_width = width // 2

# Create a new image of the appropriate size
resized = [[0 for _ in range(new_width)] for _ in range(new_height)]

# Iterate over the pixels in the resized image
for i in range(new_height):
    for j in range(new_width):
        # Compute the corresponding pixel coordinates in the original image
        pixel_x = j * 2
        pixel_y = i * 2

        # Get the pixel value from the original image and set the pixel value in the new image
        resized[i][j] = gray_image[pixel_y][pixel_x]

# Display the resized image
cv2_imshow(np.array(resized, dtype=np.uint8))
```

```python
### 6.Apply negative transformation on the image

# Create a new image of the same size with blank pixels
transformed = [[0 for i in range(width)] for j in range(height)]

for i in range(height):
    for j in range(width):
      # Store new values in the pixel
      transformed[i][j] = 255 - gray_image[i][j]

# Display the transformed image
cv2_imshow(np.array(transformed, dtype=np.uint8))
```
